```python
import pandas as pd
import numpy as np
import operator
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm



pd.options.display.float_format = '{:20,.2f}'.format
%matplotlib inline
plt.style.use('ggplot')
```

```python
# function to clean rookie df
def rookie_df_clean(data, year):
    data = data[['Rnd', 'Player', 'Age']] #removes columns not wanted
    
    x = year
    new_data = {'Rookie Year': x}
    data = data.assign(**new_data) # adding new column data rookie year
    return data
```

```python
# function to clean yearly data
def season_df_clean(data, year):
    data = data[['player', 
       'avg_depth_of_target', 'avoided_tackles', 'caught_percent',
       
        
       'grades_offense',
        'longest',
       'receptions', 'routes',
       'targeted_qb_rating', 'targets',  'yards', 'yards_after_catch',
       'yards_after_catch_per_reception', 'yards_per_reception', 'yprr']] #removes columns not wanted
    data = data.rename(columns={'player': 'Player'})
    new_data = {'ypt' : data['yards'] / data['targets'] }
    data = data.assign(**new_data) # adds new coplumn
   
    
    return data
```

```python
# function to merge rookie season year and add stats for that year.

def merge_rookieyear_stats(df_rookie, df_stats):
    df1 = pd.merge(df_rookie, df_stats, on= 'Player', how= 'left')
    return df1
```
